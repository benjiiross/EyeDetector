To find which images to send to azure with which tags, we decided to only take the tags that appear the most, as the model will struggle if there is not enough images to learn from.

We decided to take a threshold of 50 images to take from each tag. To better visualize it, we plotted the number of images per tag in a histogram.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("RFMiD_Training_Labels.csv")

plt.figure(figsize=(24, 8))

values = df.drop(["ID", "Disease_Risk"], axis=1)

sorted_values = values.sum().sort_values(ascending=False)

x = sorted_values.index
y = sorted_values.values

sns.barplot(y=y, x=x)
plt.axhline(y=50, color="r", linestyle="-")
plt.title("Number of images with each pathology", fontsize=14)
plt.xlabel("Pathology", fontsize=12)
plt.ylabel("Number of images", fontsize=12)
plt.show()
```

Removing the ID and Disease tags, we can see that the most common tags are:

- DR
- MH
- ODC
- TSLN
- DN
- MYA
- ARMD
- BRVO
- ODP
- ODE

We will use these tags to send images to Azure.
